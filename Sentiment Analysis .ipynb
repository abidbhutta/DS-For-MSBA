```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Sample data
data = {
    'Text': [
        "I love the new phone; it's fantastic!",
        "The movie was boring and uninteresting.",
        "This restaurant serves delicious food.",
        "The weather today is neither good nor bad.",
        "The service at the hotel was terrible.",
        "I'm feeling great about this project.",
        "The product meets my expectations.",
        "I can't stand the traffic in this city.",
        "Today is an average day; nothing special.",
        "The customer support team was very helpful.",
        "The book was neither good nor bad.",
        "This song is so uplifting; I love it!",
        "The computer crashed, and I lost my work.",
        "I have mixed feelings about the new design.",
        "The experience was terrible, and I'm frustrated."
    ],
    'Sentiment': [
        'Positive', 'Negative', 'Positive', 'Neutral', 'Negative',
        'Positive', 'Positive', 'Negative', 'Neutral', 'Positive',
        'Neutral', 'Positive', 'Negative', 'Neutral', 'Negative'
    ]
}

df = pd.DataFrame(data)

# Create CountVectorizer
count_vectorizer = CountVectorizer()

# Fit and transform the text data
X_tf = count_vectorizer.fit_transform(df['Text'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_tf, df['Sentiment'], test_size=0.2, random_state=42)

# Create a logistic regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest model
rf_model = RandomForestClassifier()

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions on the test set
rf_y_pred = rf_model.predict(X_test)

# Evaluate the model
rf_accuracy = accuracy_score(y_test, rf_y_pred)
rf_report = classification_report(y_test, rf_y_pred)

print(f'Random Forest Accuracy: {rf_accuracy}')
print('Random Forest Classification Report:')
print(rf_report)

```

```python
from sklearn.svm import SVC

# Create an SVM model
svm_model = SVC()

# Train the model
svm_model.fit(X_train, y_train)

# Make predictions on the test set
svm_y_pred = svm_model.predict(X_test)

# Evaluate the model
svm_accuracy = accuracy_score(y_test, svm_y_pred)
svm_report = classification_report(y_test, svm_y_pred)

print(f'SVM Accuracy: {svm_accuracy}')
print('SVM Classification Report:')
print(svm_report)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Create a Multinomial Naive Bayes model
nb_model = MultinomialNB()

# Train the model
nb_model.fit(X_train, y_train)

# Make predictions on the test set
nb_y_pred = nb_model.predict(X_test)

# Evaluate the model
nb_accuracy = accuracy_score(y_test, nb_y_pred)
nb_report = classification_report(y_test, nb_y_pred)

print(f'Naive Bayes Accuracy: {nb_accuracy}')
print('Naive Bayes Classification Report:')
print(nb_report)
```
