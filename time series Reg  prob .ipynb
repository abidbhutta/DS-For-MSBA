```python
import pandas as pd

# Assuming the data is stored in a CSV format, you can create a DataFrame using the URL or by uploading the file.
# If the data is stored in a file named 'temperature_data.csv', you can upload it to Colab first.

# Option 1: Load from a URL if the data is publicly accessible
df = pd.read_csv('/content/drive/MyDrive/Tempreture_1901_2016_Pakistan.csv')

# Option 2: Upload the file directly to Colab and then load it
# Uncomment the following lines if you have uploaded the file to Colab
# from google.colab import files
# uploaded = files.upload()
# df = pd.read_csv(io.StringIO(uploaded['temperature_data.csv'].decode('utf-8')))

# Display the first few rows of the DataFrame
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the data into a DataFrame
# Assuming you've already loaded the data into the 'df' DataFrame

# Display the first few rows of the DataFrame
print(df.head())

# Trim any extra whitespaces from column names
df.columns = df.columns.str.strip()

# Extract features (independent variables) and target (dependent variable)
X = df[['Year', 'Month']]
y = df['Temperature - (Celsius)']

# Convert categorical variables (Month) to numerical using one-hot encoding
X_encoded = pd.get_dummies(X, columns=['Month'], drop_first=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Plot actual vs predicted values
plt.scatter(X_test['Year'], y_test, color='black', label='Actual')
plt.scatter(X_test['Year'], y_pred, color='red', label='Predicted')
plt.xlabel('Year')
plt.ylabel('Temperature - (Celsius)')
plt.legend()
plt.show()

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the data into a DataFrame
# Assuming you've already loaded the data into the 'df' DataFrame

# Display the first few rows of the DataFrame
print(df.head())

# Trim any extra whitespaces from column names
df.columns = df.columns.str.strip()

# Extract features (independent variables) and target (dependent variable)
X = df[['Year', 'Month']]
y = df['Temperature - (Celsius)']

# Convert categorical variables (Month) to numerical using one-hot encoding
X_encoded = pd.get_dummies(X, columns=['Month'], drop_first=True)

# Filter data for the year 2021 (or any other available year)
data_2021 = df[df['Year'] == 2021]

# Check if there are any samples for the year 2021
if not data_2021.empty:
    X_2021 = data_2021[['Year', 'Month']]
    y_2021 = data_2021['Temperature - (Celsius)']

    # Convert categorical variables (Month) to numerical using one-hot encoding for 2021 data
    X_encoded_2021 = pd.get_dummies(X_2021, columns=['Month'], drop_first=True)

    # Ensure that the columns in X_encoded_2021 are in the same order as X_encoded
    X_encoded_2021 = X_encoded_2021.reindex(columns=X_encoded.columns, fill_value=0)

    # Create a linear regression model
    model = LinearRegression()

    # Train the model
    model.fit(X_encoded, y)

    # Make predictions for the year 2021
    y_pred_2021 = model.predict(X_encoded_2021)

    # Plot the data for the year 2021 along with the regression line
    plt.figure(figsize=(10, 6))

    # Plot actual data
    plt.scatter(X_2021['Month'], y_2021, label='Actual Data', color='blue')
    # Plot predictions
    plt.scatter(X_2021['Month'], y_pred_2021, label='Predictions', color='red', marker='x')

    # Plot regression line
    plt.plot(X_2021['Month'], y_pred_2021, label='Regression Line', color='orange')

    plt.xlabel('Month')
    plt.ylabel('Temperature - (Celsius)')
    plt.title('Temperature Prediction for the Year 2021 with Regression Line')
    plt.legend()
    plt.show()
else:
    print("No data available for the year 2021.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the data into a DataFrame
# Assuming you've already loaded the data into the 'df' DataFrame

# Display the first few rows of the DataFrame
print(df.head())

# Trim any extra whitespaces from column names
df.columns = df.columns.str.strip()

# Extract features (independent variables) and target (dependent variable)
X = df[['Year', 'Month']]
y = df['Temperature - (Celsius)']

# Convert categorical variables (Month) to numerical using one-hot encoding
X_encoded = pd.get_dummies(X, columns=['Month'], drop_first=True)

# Choose the year for visualization
selected_year = 2018

# Filter data for the selected year
data_selected_year = df[df['Year'] == selected_year]

# Check if there are any samples for the selected year
if not data_selected_year.empty:
    X_selected_year = data_selected_year[['Year', 'Month']]
    y_selected_year = data_selected_year['Temperature - (Celsius)']

    # Convert categorical variables (Month) to numerical using one-hot encoding for selected year data
    X_encoded_selected_year = pd.get_dummies(X_selected_year, columns=['Month'], drop_first=True)

    # Ensure that the columns in X_encoded_selected_year are in the same order as X_encoded
    X_encoded_selected_year = X_encoded_selected_year.reindex(columns=X_encoded.columns, fill_value=0)

    # Create a linear regression model
    model = LinearRegression()

    # Train the model
    model.fit(X_encoded, y)

    # Make predictions for the selected year
    y_pred_selected_year = model.predict(X_encoded_selected_year)

    # Plot the data for the selected year along with the regression line
    plt.figure(figsize=(10, 6))

    # Plot actual data
    plt.scatter(X_selected_year['Month'], y_selected_year, label=f'Actual Data ({selected_year})', color='blue')
    # Plot predictions
    plt.scatter(X_selected_year['Month'], y_pred_selected_year, label='Predictions', color='red', marker='x')

    # Plot regression line
    plt.plot(X_selected_year['Month'], y_pred_selected_year, label='Regression Line', color='orange')

    plt.xlabel('Month')
    plt.ylabel('Temperature - (Celsius)')
    plt.title(f'Temperature Prediction for the Year {selected_year} with Regression Line')
    plt.legend()
    plt.show()
else:
    print(f"No data available for the year {selected_year}.")
```
