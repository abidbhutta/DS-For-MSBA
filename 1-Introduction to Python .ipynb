**Activity 3: Using Variables and Assign Statements**

* Write a script or program that given **distance in kilometers** and **given  time in hours**, to calculate and print out speed in **kilometers per hour**, speed **miles per hour**, and speed **meters per second.**

* Here are some hints:

  * The formula for calculating speed is  (**speed = distance/time**).
  * To convert kilometers to miles, divide the kilometers by 1.6.
  * To convert kilometers to meters, multiply the kilometers by 1,000.
  * To convert hours to seconds, multiply hours by 3,600.

````
# distance in kilometers
distance_in_km = 100


# time in hours
time_in_hours = 5

# convert distance into miles
distance_in_miles = distance_in_km / 1.6

# convert distance into meters
distance_in_meters = distance_in_km * 1000

# convert time in seconds
time_in_seconds = time_in_hours * 3600

# speed in kilometrs per hour
speed_in_kph = distance_in_km / time_in_hours

# speed in miles per hour
speed_in_mph = distance_in_miles / time_in_hours

# speed in meters per second.
speed_in_mps = distance_in_meters / time_in_seconds

# Now we shall print out our results
print("Speed in kph: ", speed_in_kph)
print("Speed in mph: ", speed_in_mph)
print("Speed in mps: ", speed_in_mps)
````

* Write a script that will calculate the area and circumference of a circle with a radius of 7. In this activity, we shall get       better acquainted with variable assignment as well as variable naming conventions.

* Here are some hints:

  * The formula for calculating area is π * r², r being the radius.
  * The formula for calculating circumference is 2 * π * r.

````
# The Solution of Activity No. 4

# radius of circle
radius_of_circle = 7

# constant value PI
PI = 3.14159

# area of circle
area_of_circle = PI * radius_of_circle * radius_of_circle

# circumference of circle
circumference_of_circle = 2 * PI * radius_of_circle

# print the area and circumference of a circle
print("Area of circle : ", area_of_circle)
print("Circumference of circle: ", circumference_of_circle)
````

**calculator**

````
def calculate(a, b, operator):
    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '*':
        return a * b
    elif operator == '/':
        if b != 0:
            return a / b
        else:
            return "Error: Division by zero is not allowed."
    else:
        return "Error: Invalid operator."

print(":::: Calculator ::::")

while True:
    a = float(input("Enter value of a: "))
    b = float(input("Enter value of b: "))
    operator = input("Enter the operation (+, -, *, /) or 'exit' to end: ")

    if operator.lower() == 'exit':
        print("Exiting the calculator.")
        break

    result = calculate(a, b, operator)
    print(f"Answer: {result}")
````

Reverse a number

````
def reverse_number(number):
    reversed_number = 0

    while number > 0:
        digit = number % 10
        reversed_number = (reversed_number * 10) + digit
        number = int(number / 10)  # Perform floor division without using //

    return reversed_number

# Input
number = int(input("Enter a number: "))

# Output
result = reverse_number(number)
print("Reversed Number:", result)
````

Assignment : Reverse a sentense

````
def reverse_sentence(sentence):
    words = sentence.split()
    reversed_words = []

    for word in reversed(words):
        reversed_words.append(word)

    reversed_sentence = ' '.join(reversed_words)
    return reversed_sentence

# Input
sentence = input("Enter a sentence: ")

# Output
result = reverse_sentence(sentence)
print("Reversed Sentence:", result)
````

````
def reverse_sentence(sentence):
    words = sentence.split()
    reversed_sentence = ' '.join(reversed(words))
    return reversed_sentence

# Input
sentence = input("Enter a sentence: ")

# Output
result = reverse_sentence(sentence)
print("Reversed Sentence:", result)
````
